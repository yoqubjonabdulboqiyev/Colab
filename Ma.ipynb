5-bosqich: NASA Climate Data â€“ Harorat oâ€˜zgarishi (demo maâ€™lumot)

```python
data_temp = {
    1880: -0.2, 1900: -0.1, 1920: 0.0, 1940: 0.1,
    1960: 0.0, 1980: 0.3, 2000: 0.6, 2010: 0.9, 2020: 1.2, 2024: 1.3
}
df_temp = pd.DataFrame.from_dict(data_temp, orient="index", columns=["Temp_Anomaly"])
df_temp.index.name = "Year"
df_temp.head()
```

6-bosqich: Harorat grafigi

```python
plt.figure(figsize=(10,6))
plt.plot(df_temp.index, df_temp["Temp_Anomaly"], marker='o', color="red")
plt.title("Global oâ€˜rtacha harorat anomaliyalari (1880â€“2024, NASA GISS)")
plt.xlabel("Yil")
plt.ylabel("Â°C (bazaviy davrga nisbatan)")
plt.grid(True)
plt.show()
```

```python
data_co2 = {
    1960: 317, 1980: 338, 2000: 370, 2010: 390, 2020: 414, 2024: 421
}
df_co2 = pd.DataFrame.from_dict(data_co2, orient="index", columns=["CO2_ppm"])
df_co2.index.name = "Year"
df_co2
```

10-bosqich: IPCC prognozlari (RCP ssenariylar)

```python
scenarios = {
    "RCP2.6": [1.5, 1.6, 1.7, 1.8],
    "RCP4.5": [1.6, 2.0, 2.4, 2.6],
    "RCP8.5": [1.8, 2.7, 3.8, 4.5]
}
years = [2030, 2050, 2075, 2100]
df_ipcc = pd.DataFrame(scenarios, index=years)
df_ipcc
```

11-bosqich: Prognoz grafiklari

```python
plt.figure(figsize=(10,6))
for col in df_ipcc.columns:
    plt.plot(df_ipcc.index, df_ipcc[col], marker='o', label=col)
plt.title("IPCC prognozlari â€“ Global harorat oâ€˜zgarishi (Â°C, 2030â€“2100)")
plt.xlabel("Yil")
plt.ylabel("Harorat (Â°C)")
plt.legend()
plt.grid(True)
plt.show()
```
